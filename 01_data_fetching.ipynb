# Step 1: Data Fetching
In this notebook, we'll download historical market data using `yfinance`. This is the first step in building any trading algorithm.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Set a style for prettier graphs
plt.style.use('seaborn-v0_8-darkgrid')
```

## 1. Fetching Data for the S&P 500 ETF (SPY)
We'll fetch daily data from the last 2 years.

```python
ticker = "SPY"
start_date = "2022-01-01"
end_date = "2024-01-01"

# Download the data
df = yf.Ticker(ticker).history(start=start_date, end=end_date)

# Show the first few rows
df.head()
```

## 2. Visualizing the Data
Let's plot the Close price to see the trend.

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label=f'{ticker} Close Price')
plt.title(f"{ticker} Historical Close Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```
